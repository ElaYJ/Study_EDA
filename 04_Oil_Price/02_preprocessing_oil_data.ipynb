- ν•„μ”ν• μ»¬λΌ λ°μ΄ν„°λ§ μ¶”λ ¤μ„ μƒλ΅μ΄ DataFrame λ§λ“¤κΈ°

```python
stations = pd.DataFrame({
	"μƒνΈ": stations_raw["μƒνΈ"],
	"μ£Όμ†": stations_raw["μ£Όμ†"],
	"κ°€κ²©": stations_raw["νλ°μ "],
	"μ…€ν”„": stations_raw["μ…€ν”„μ—¬λ¶€"],
	"μƒν‘": stations_raw["μƒν‘"]
})
stations.tail()
```

- "μ£Όμ†"μ—μ„ κµ¬ μ΄λ¦„ μ¶”μ¶ν•κΈ°

```python
for each_address in stations["μ£Όμ†"]:
    print(each_address.split())
```

```python
for each_address in stations["μ£Όμ†"]:
    print(each_address.split()[1])
```

```python
# means_1.
stations["κµ¬"] = [each_address.split()[1] for each_address in stations["μ£Όμ†"]]
stations
```

```python
# means_2.
stations["Gu"] = stations["μ£Όμ†"].apply(lambda addr: addr.split()[1])
stations
```

```python
# case_1. κ°€κ²© μ •λ³΄κ°€ μλ” μ£Όμ μ†λ§ μ‚¬μ©
stations = stations[stations["κ°€κ²©"] != "-"]
```

```python
# case_2. '-'μ„ NaNμΌλ΅ λ°”κΎΌ ν›„ NaNλ§ μ§€μ°κΈ°
stations[["κ°€κ²©"]] = stations[["κ°€κ²©"]].applymap(lambda val: np.nan if val == '-' else val)
stations.tail()
```
