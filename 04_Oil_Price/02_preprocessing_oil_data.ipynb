- 파일명 저장

```python
oil_stations_files = glob("./result_data/지역_*.xls")
oil_stations_files[:5]
```

### 🔰 DataFrame 생성

- 엑셀 파일 읽기

	- 일단 하나만 읽어본다.

	- 컬럼이 원하는 형태로 나오지 않는다.

	- 엑셀 파일을 열어보면 위에 두개의 빈 rows를 확인할 수 있다.

	- 읽어올 때 제외하고 읽어온다.

```python
pd.read_excel(oil_stations_files[0]).tail(2)
```

```python
tmp = pd.read_excel(oil_stations_files[0], header=2)
tmp.tail(2)
```

```python
# 구별 전체 엑셀 파일 읽어오기

tmp_gu_df_raw = []

for file_name in oil_stations_files:
    tmp_gu_df = pd.read_excel(file_name, header=2)
    tmp_gu_df_raw.append(tmp_gu_df)
    
tmp_gu_df_raw
```

- `concat()`

	- concatenation, 연결·연쇄·연속
	
	- DataFrame의 구조와 형식이 동일하면 연달아 이어 붙이기가 가능하다.

```python
stations_raw = pd.concat(tmp_gu_df_raw)
stations_raw
```

👆 여러 개의 DataFrame을 합치면서 인덱스값이 제대로 표시되지 않음.

- 인덱스를 다시 재정렬 해준다.

```python
stations_raw.reset_index(inplace=True)
stations_raw.tail()
```

```python
stations_raw.info()
```

### 🔰 DataFrame 편집

```python
stations_raw.columns
```

- 필요한 컬럼 데이터만 추려서 새로운 DataFrame 만들기

```python
stations = pd.DataFrame({
	"상호": stations_raw["상호"],
	"주소": stations_raw["주소"],
	"가격": stations_raw["휘발유"],
	"셀프": stations_raw["셀프여부"],
	"상표": stations_raw["상표"]
})
stations.tail()
```

- "주소"에서 구 이름 추출하기

```python
for each_address in stations["주소"]:
    print(each_address.split())
```

```python
for each_address in stations["주소"]:
    print(each_address.split()[1])
```

```python
# means_1.
stations["구"] = [each_address.split()[1] for each_address in stations["주소"]]
stations
```

```python
# means_2.
stations["Gu"] = stations["주소"].apply(lambda addr: addr.split()[1])
stations
```

```python
stations["구"].unique(), len(stations["구"].unique())
```

- `astype()` : 데이터 형변환

	- "가격" 데이터 변환: object --> float

```python
stations.info()
```

```python
stations["가격"] = stations["가격"].astype("float")
```

👆 "가격" value에 존재하는 '-'를 float으로 형변환 할 수 없어 오류 발생

- 가격 정보 없는 주유소 확인

```python
stations[stations["가격"] == "-"]
```

```python
# case_1. 가격 정보가 있는 주유소만 사용
stations = stations[stations["가격"] != "-"]
```

```python
# case_2. '-'을 NaN으로 바꾼 후 NaN만 지우기
stations[["가격"]] = stations[["가격"]].applymap(lambda val: np.nan if val == '-' else val)
stations.tail()
```

```python
stations.dropna(axis=0).tail()
```

```python
stations.dropna(axis=0, inplace=True)
stations.tail()
```

```python
stations[stations["가격"] == "-"]
```

```python
stations["가격"] = stations["가격"].astype("float")
```
