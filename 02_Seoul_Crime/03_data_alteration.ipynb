```python
import pandas as pd

crime_raw_data = pd.read_csv("./result_data/01_crime_raw_data.csv")
crime_raw_data.head()
```

#### ğŸ”° DataFrame ì¬í¸ì„±

> raw_dataë¥¼ pivot_table()ì„ ì‚¬ìš©í•´ DataFrameì„ ì¬í¸ì„±í•˜ê³  "ê±´ìˆ˜"ë¥¼ í•©ìœ¼ë¡œ ì¬ì§‘ê³„í•œë‹¤.

```python
import numpy as np

crime_station = crime_raw_data.pivot_table(
	index="êµ¬ë¶„",
	columns=["ì£„ì¢…", "ë°œìƒê²€ê±°"],
	aggfunc=np.sum
)

crime_station.head()
```

```python
crime_station["êµ¬ë³„"] = np.nan
crime_station["lat"] = np.nan # latitude, ìœ„ë„
crime_station["lng"] = np.nan # longitude, ê²½ë„
crime_station.head()
```

#### ğŸ”° ìƒˆë¡œìš´ ë°ì´í„° ì¶”ê°€

> ê²½ì°°ì„œ ì´ë¦„ìœ¼ë¡œ googlemapsì—ì„œ ìœ„ì¹˜ ì •ë³´ë¥¼ ì–»ì–´ì™€ í•„ìš”í•œ ë°ì´í„°ë¡œ ìƒˆë¡œ ì¶”ê°€í•œ ì»¬ëŸ¼ valuesë¥¼ ì±„ìš´ë‹¤.

```python
import googlemaps

gmaps = googlemaps.Client(key="")
```

```python
crime_station.index
```

```python
cnt = 0
for idx, row in crime_station.iterrows():
    if idx == "ë™ì‘":
        station_name = idx + "ê²½ì°°ì„œ"
    else:
        station_name = "ì„œìš¸" + idx + "ê²½ì°°ì„œ"
    
    temp = gmaps.geocode(station_name, language="ko")
    
    # ì£¼ì†Œ ì •ë³´ ì–»ì–´ì˜¤ê¸°
    tmp_gu = temp[0].get("formatted_address")
    
    # ìœ„ì¹˜(ìœ„ë„, ê²½ë„) ê°€ì ¸ì˜¤ê¸°
    lat = temp[0].get("geometry")["location"]["lat"]
    lng = temp[0].get("geometry")["location"]["lng"]
    
    # ê° ì»¬ëŸ¼ì— ê°’ ì±„ìš°ê¸°
    crime_station.loc[idx, "êµ¬ë³„"] = tmp_gu.split()[2]
    crime_station.loc[idx, "lat"] = lat
    crime_station.loc[idx, "lng"] = lng
    
    print(cnt)
    cnt += 1
```
