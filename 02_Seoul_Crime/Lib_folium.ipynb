### ğŸ”° folium.Circle(), folium.CircleMarker()

````python
state_data = pd.read_csv("../data/02. US_Unemployment_Oct2012.csv")
state_data.tail(2)
````

````python
m = folium.Map([43, -102], zoom_start=3)

folium.Choropleth(
	geo_data="../data/02. us-states.json", # ê° ì£¼ë§ˆë‹¤ì˜ ê²½ê³„ì„  ì¢Œí‘œê°’ì´ ë‹´ê¸´ ë°ì´í„°
	data=state_data, # Series or Pandas DataFrame í˜•íƒœê°€ ë“¤ì–´ê°€ì•¼ í•œë‹¤.
	columns=["State", "Unemployment"], # DataFrame columns
	key_on="feature.id",
	fill_color="YlGn", #"BuPu",
	fill_opacity=0.5, # 0~1
	line_opacity=0.2, # 0~1
	legend_name="Unemployment rate (%)"
).add_to(m)

m
````
