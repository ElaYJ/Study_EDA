### 🔰 folium.Marker()

- 지도에 마커 생성

- popup : 마우스로 클릭하면 창이 popup

- tooltip : 마우스를 마커에 가져가면 나타남.

````python
fmap = folium.Map(
	location=[37.544564958079896, 127.05582307754338], # 성수역
	zoom_start=14,
 	tiles="OpenStreetMap"
)

# 마커 생성
folium.Marker((37.544564958079896, 127.05582307754338)).add_to(fmap)

# popup
folium.Marker(
	location=[37.54712311308356, 127.04721916917774], # 뚝섬역
	popup="<b>Subway</b>"
).add_to(fmap)

# tooltip
folium.Marker(
	location=[37.544564958079896, 127.05582307754338],
	popup="<b>Subway</b>",
	tooltip="<i>성수역</i>"
).add_to(fmap)

# html 문법 사용 가능
folium.Marker(
	location=[37.54558642069953, 127.05729705810472],
	popup="<a href='https://zero-base.co.kr/' target=_'blink'>제로베이스</a>", # target=_'blink' : 새 창에서 링크가 뜬다.
	tooltip="<i>Zerobase</i>"
).add_to(fmap)

fmap
````

### 🔰 folium.ClickForMarker()

- 지도위에 마우스로 클릭했을 때 마커를 생성해준다

- 마커를 클릭하면 팝업창이 뜨게 해준다.

````python
m = folium.Map(
    location=[37.544564958079896, 127.05582307754338], # 성수역 
    zoom_start=14,
    tiles="OpenStreetMap"
)

m.add_child(folium.ClickForMarker(popup="ClickForMarker"))
````

### 🔰 folium.Circle(), folium.CircleMarker()

- Circle : 절대적 크기

- CircleMarker : 지도 배율에 상대적 크기

````python
m = folium.Map(
    location=[37.544564958079896, 127.05582307754338], # 성수역 
    zoom_start=14,
    tiles="OpenStreetMap"
)

folium.Circle(
	location=[37.555243442409406, 127.04370422643919], # 한양대학교
	radius=100,
	fill=True,
	color="#eb9e34", # colour picker 검색해서 색상값을 가져올 수 있다.
	fill_color="red",
	popup="Circle Popup",
	tooltip="Circle Tooltip"
).add_to(m)

folium.CircleMarker(
	location=[37.54347089498245, 127.04439204503049],
	radius=100,
	fill=True,
	color="#34ebc6",
    fill_color="#c634eb",
    popup="CircleMarker Popup",
    tooltip="CircleMarker Tooltip"
).add_to(m)

m
````

### 🔰 folium.Choropleth

- 지도에 colormap을 표현한다.

- json 파일을 이용해 경계선과 id를 각 지역에 구현할 수 있다.

- csv 파일에 이에 매칭되는 id와 값을 가지도록 함.

````python
import json
````

````python
state_data = pd.read_csv("../data/02. US_Unemployment_Oct2012.csv")
state_data.tail(2)
````

````python
m = folium.Map([43, -102], zoom_start=3)

folium.Choropleth(
	geo_data="../data/02. us-states.json", # 각 주마다의 경계선 좌표값이 담긴 데이터
	data=state_data, # Series or Pandas DataFrame 형태가 들어가야 한다.
	columns=["State", "Unemployment"], # DataFrame columns
	key_on="feature.id",
	fill_color="YlGn", #"BuPu",
	fill_opacity=0.5, # 0~1
	line_opacity=0.2, # 0~1
	legend_name="Unemployment rate (%)"
).add_to(m)

m
````

### 🔰 아파트 유형 지도 시각화 

- 공공데이터포털

````python
df = pd.read_csv("../data/02. 서울특별시 동작구_주택유형별 위치 정보 및 세대수 현황_20210825.csv", encoding="cp949")
df.tail(2)
````

````python
df.info()
````

👆 위도, 경도 값이 4개 빠져있다.

- 위도, 경도가 NaN인 데이터 제거

````python
df = df.dropna()
df.info()
````

````python
df
````

````python
df.reset_index() # default: drop=False -> index 생성
````

````python
df.reset_index(drop=True) # drop=True -> index 삭제
````

- 인덱스랑 연번이 맞지 않아 연번 컬럼을 삭제하려고 하니 오류가 발생한다.

````python
del df["연번"]
````

- 컬럼 데이터를 확인해 보니 "연번"이랑 "분류" 뒤에 한 칸 공백문자가 들어가 있다.

````python
df.columns
````

- 컬럼명 수정

````python
df = df.rename(columns={"연번 ": "연번", "분류 ": "분류"})
df.연번[:10]
````

````python
del df["연번"]
````

````python

df.tail(2)
````

````python
df.위도, df.경도
````

````python
df.describe()
````

````python
# folium 

m = folium.Map(location=[37.50589466533131, 126.93450729567374], zoom_start=15) # 동작구

for idx, rows in df.iterrows():
    
    # location 
    lat, lng = rows.위도, rows.경도
    
    # Marker 
    folium.Marker(
        location=[lat, lng],
        popup=rows.주소,
        tooltip=rows.분류, 
        icon=folium.Icon(
            icon="home",
            color="lightred" if rows.세대수 >= 199 else "lightblue",
            icon_color="darkred" if rows.세대수 >= 199 else "darkblue",
        )
    ).add_to(m)

    # Circle
    folium.Circle(
        location=[lat, lng],
        radius=rows.세대수 * 0.5, # 세대수가 커지면 원크기도 같이 증가
        fill=True, 
        color="pink" if rows.세대수 >= 518 else "green",
        fill_color="pink" if rows.세대수 >= 518 else "green",
    ).add_to(m)
    
m   
````
