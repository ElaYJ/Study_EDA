```python
num = ["ê°•ê°„ê²€ê±°", "ê°•ë„ê²€ê±°", "ì‚´ì¸ê²€ê±°", "ì ˆë„ê²€ê±°", "í­ë ¥ê²€ê±°"]
den = ["ê°•ê°„ë°œìƒ", "ê°•ë„ë°œìƒ", "ì‚´ì¸ë°œìƒ", "ì ˆë„ë°œìƒ", "í­ë ¥ë°œìƒ"]

crime_anal_gu[num].div(crime_anal_gu[den].values).head()
```
