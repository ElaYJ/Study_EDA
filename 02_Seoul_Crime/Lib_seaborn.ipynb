### üî∞ ÏòàÏ†ú1. seaborn Í∏∞Ï¥à

```python
np.linspace(0, 14, 100)
```

```python
x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2 * np.sin(x + 0.5)
y3 = 3 * np.sin(x + 1.0)
y4 = 4 * np.sin(x + 1.5)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4) # plotÏùÄ Ïã§ÏÑ† Í∑∏ÎûòÌîÑÏù¥Îã§.
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
sns.despine(offset=10)
plt.show()
```

- sns.set_sytle()

	- "white", "whitegrid", "ticks", "dark", "darkgrid"

```python
def drawGraph():
	plt.figure(figsize=(10, 6))
	plt.plot(x, y1, x, y2, x, y3, x, y4)
	plt.show()
```

```python
sns.set_style("white")
drawGraph()
```

```python
sns.set_style("whitegrid")
drawGraph()
```

```python
sns.set_style("ticks")
drawGraph()
```

```python
sns.set_style("dark")
drawGraph()
```

```python
sns.set_style("darkgrid")
drawGraph()
```

### üî∞ ÏòàÏ†ú2. seaborn tips data

> seabornÏóêÎäî Ïã§ÏäµÏö© Îç∞Ïù¥ÌÑ∞Í∞Ä Î™á Í∞ú ÎÇ¥Ïû•ÎêòÏñ¥ ÏûàÎã§. Í∑∏ Ï§ë ÌïòÎÇòÏù∏ 'tips'Î•º Î∂àÎü¨Ïò®Îã§.

- boxplot</br>
- swarmplot</br>
- lmplot</br>

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips.info()
```

#### üé´ boxplot

- boxplot Í∑∏Î†§Î≥¥Í∏∞

```python
tips[tips["total_bill"] > 40]
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=tips["total_bill"])
plt.show()
```

- boxplotÏóê Ïª¨Îüº ÏßÄÏ†ï

```python
tips["day"].unique()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

- boxplotÏùò `hue`, `palette` param

	- hue : Ïπ¥ÌÖåÍ≥†Î¶¨ Îç∞Ïù¥ÌÑ∞ ÌëúÌòÑ
	- palette : Set 1 ~ 3

```python
tips.head(2)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker', palette="Set3")
plt.show()
```

#### üé´ swarmplot

- color: 0 ~ 1 ÏÇ¨Ïù¥ Í∞í,  Í≤ÄÏùÄÏÉâ(0)Î∂ÄÌÑ∞ Ìù∞ÏÉâ(1)ÍπåÏßÄ Ï°∞Ï†à 

```python
plt.figure(figsize=(8, 6))
sns.swarmplot(x="day", y="total_bill", data=tips, color=".5")
plt.show()
```

#### üé´ boxplot with swarmplot

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color="0.25")
plt.show()
```

#### üé´ lmplot

- total_billÍ≥º tip ÏÇ¨Ïù¥Ïùò Í¥ÄÍ≥Ñ ÌååÏïÖÌï† Ïàò ÏûàÎã§.

- Í∑∏Î¶ºÏûê ÏòÅÏó≠Ïù¥ Ï¢ÅÏùÑÏàòÎ°ù ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÍ∞Ä ÎÜíÎã§.

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, height=7) # size => heightÎ°ú Î≥ÄÍ≤Ω
plt.show()
```

- `hue` param ÏÇ¨Ïö©

```python
sns.lmplot(x="total_bill", y="tip", data=tips, height=7, hue="smoker")
plt.show()
```

### üî∞ ÏòàÏ†ú3. flights data

- heatmap

```python
flights = sns.load_dataset("flights")
flights
```

```python
flights.info()
```

- pivot(index, columns, values)

	- pivot() Ìï®ÏàòÎ•º ÏÇ¨Ïö©Ìï¥ DataFrameÏùÑ Ïû¨Ìé∏ÏÑ±ÌïúÎã§.

```python
flights = flights.pivot(index="month", columns="year", values="passengers")
flights
```

#### üé´ heatmap

- heatmapÏùÑ Ïù¥Ïö©ÌïòÎ©¥ Ï†ÑÏ≤¥ Í≤ΩÌñ•ÏùÑ Ïïå Ïàò ÏûàÎã§.

- Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ ÌùêÎ¶ÑÏùÑ Ïûò Î≥º Ïàò ÏûàÎã§.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data=flights, annot=True, fmt="f") # annot=True Îç∞Ïù¥ÌÑ∞ Í∞í ÌëúÏãú
plt.show()
```

- Îã§Î•∏ colormap ÏÇ¨Ïö©

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data=flights, annot=True, fmt="d", cmap="YlGnBu") # fmt="d" Ï†ïÏàòÌòï ÌëúÌòÑ
plt.show()
```

- Îç∞Ïù¥ÌÑ∞ Ìï¥ÏÑù

	- Ïó∞ÎèÑÍ∞Ä Ï¶ùÍ∞ÄÌï†ÏàòÎ°ù Ï†ÑÏ≤¥ ÏäπÍ∞ùÏàòÎèÑ Ï¶ùÍ∞Ä
	- Ïó∞ÎèÑÎ≥ÑÎ°ú Ïó¨Î¶ÑÏóê ÏäπÍ∞ùÏàòÍ∞Ä ÎßéÏïòÎã§.
	- Ïó∞ÎèÑÍ∞Ä ÏßÄÎÇ†ÏàòÎ°ù Í≤®Ïö∏Ï≤† ÏäπÍ∞ùÏù¥ Ï¶ùÍ∞ÄÌïòÎ†§Í≥† ÌïòÍ≥† ÏûàÎã§.

### üî∞ ÏòàÏ†ú5. anscombe data

- lmplot

```python
anscombe = sns.load_dataset("anscombe")
anscombe.tail()
```

```python
anscombe["dataset"].unique()
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'I'"), ci=None, height=7)
# ci= Ïã†Î¢∞Íµ¨Í∞Ñ(confidence interval) ÏÑ†ÌÉù, NoneÏùÄ Ïã†Î¢∞Íµ¨Í∞Ñ ÏòÅÏó≠Ïù¥ Î≥¥Ïù¥ÏßÄ ÏïäÍ≤å ÏÑ§Ï†ïÌïúÎã§.
plt.show()
```

- ÎßàÏª§ ÏÇ¨Ïù¥Ï¶à Î≥ÄÍ≤Ω

```python
sns.lmplot(
	x="x", y="y",
	data=anscombe.query("dataset == 'I'"),
	ci=None,
	height=7,
	scatter_kws={"s": 70} # ÎßàÏª§ÌÅ¨Í∏∞ ÏÑ§Ï†ï
)
plt.show()
```

- `order` param

```python
sns.lmplot(
	x="x", y="y",
	data=anscombe.query("dataset == 'II'"), # 2Ï∞®Ïãù
	order=1, # Ï∞®ÏàòÏóê Îî∞Îùº Ïù∏Ïàò Î≥ÄÍ≤Ω
	ci=None,
	height=7,
	scatter_kws={"s": 80}
)
plt.show()
```

```python
sns.lmplot(
	x="x", y="y",
	data=anscombe.query("dataset == 'II'"), # 2Ï∞®Ïãù
	order=2, # Ï∞®ÏàòÏóê Îî∞Îùº Ïù∏Ïàò Î≥ÄÍ≤Ω
	ci=None,
	height=7,
	scatter_kws={"s": 80}
)
plt.show()
```

- `robust` param

- Í≤ΩÌñ•ÏóêÏÑú ÎßéÏù¥ Î≤óÏñ¥ÎÇú Í∞í(outlier) ÎïåÎ¨∏Ïóê Í∑∏Î¶¨ÌîÑÍ∞Ä ÏúÑ, ÏïÑÎûòÎ°ú ÎÅåÎ†§ Í∞àÏàò ÏûàÎã§.

```python
sns.lmplot(
	x="x", y="y",
	data=anscombe.query("dataset == 'III'"), # outlierÍ∞Ä Ï°¥Ïû¨ÌïòÎäî Îç∞Ïù¥ÌÑ∞
	ci=None,
	height=7,
	scatter_kws={"s": 80}
)
plt.show()
```

```python
sns.lmplot(
	x="x", y="y",
	data=anscombe.query("dataset == 'III'"), # outlierÍ∞Ä Ï°¥Ïû¨ÌïòÎäî Îç∞Ïù¥ÌÑ∞
	robust=True, # Í≤ΩÌñ•ÏóêÏÑú ÎßéÏù¥ Îñ®Ïñ¥ÏßÑ Îç∞Ïù¥ÌÑ∞Î•º Î¨¥ÏãúÌïúÎã§.
	ci=None,
	height=7,
	scatter_kws={"s": 80}
)
plt.show()
```
