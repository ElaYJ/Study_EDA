```python
import pandas as pd

# .txt 파일을 pandas.read_csv()로 읽는다.
df = pd.read_csv("./data/report.txt", sep="\t")
df.head()
```

```python
# thousands=',' 옵션이 안먹힘.
df = pd.read_csv("./data/report.txt", sep="\t", thousands=',')
df.tail()
```

```python
df.drop([0,1,2], inplace=True) #--> Row가 삭제됨. --> default axis=0 이여서 Row가 삭제되었던 것이었음.
df.head()
```

```python
df.drop(labels=['기간', '자치구'], axis=1) #--> axis=1 이면 Column이 삭제됨.
```

```python
# df.reset_index(inplace=True)
# del df["index"]
# 대신
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
# 모범답안
new_columns = ['기간', '자치구', '세대', '합계', '남자', '여자', '한국인 계', '한국인 남자', '한국인 여자', '등록외국인 계', '등록외국인 남자', '등록외국인 여자', '세대당인구', '65세이상고령자']
df.columns = new_columns
```

```python
# 컬럼명 변경
df.rename(
	columns={
		"인구": "합계",
		"인구.1": "남자",
		"인구.2": "여자",
		"인구.3": "한국인 계",
		"인구.4": "한국인 남자",
		"인구.5": "한국인 여자",
		"인구.6": "등록외국인 계",
		"인구.7": "등록외국인 남자",
		"인구.8": "등록외국인 여자",
	},
	inplace=True
)
df.columns
```

```python
col_list = df.columns.to_list()
col_list
```

```python
df.head()
```

```python
# Can only use .str accessor with string values!
df["세대"].str.replace(",", "").astype("int64")
```

```python
df["세대당인구"].astype("float")
```

```python
# 모범 답안
target_cols = ['세대', '합계', '남자', '여자', '한국인 계', '한국인 남자', '한국인 여자', '등록외국인 계', '등록외국인 남자', '등록외국인 여자', '65세이상고령자']
df[target_cols] = df[target_cols].apply(axis=1, func=lambda col: col.str.replace(',', '').astype(int))
df['세대당인구'] = df['세대당인구'].astype(float)
```

```python
for column in col_list[2:]:
    if column == '세대당인구':
        df[column] = df[column].astype("float")
    else:
        df[column] = df[column].str.replace(",", "").astype("int64")
df.head()
```

```python
df.info()
```

```python
region_dict = {'도심권': ['종로구', '중구', '용산구'],
               '동북권': ['성동구', '광진구', '동대문구', '중랑구', '성북구', '강북구', '도봉구', '노원구'],
               '서북권': ['은평구', '서대문구', '마포구'],
               '서남권': ['양천구', '강서구', '구로구', '금천구', '영등포구', '동작구', '관악구'],
               '동남권': ['서초구', '강남구', '송파구', '강동구']
               }
```

```python
# Key와 Value를 뒤집음.
gu_dict = {gu: region for region, gu_list in region_dict.items() for gu in gu_list}
gu_dict
```

```python
# 모범 답안

gu_dict = {gu: region for region, gu_list in region_dict.items() for gu in gu_list}

for idx, row in df.iterrows():
    df.loc[idx, '권역'] = gu_dict[row['자치구']]
```

```python
for idx, row in df.iterrows():
    if row["자치구"] in region_dict["도심권"]:
        df.loc[idx, "권역"] = "도심권"
    elif row["자치구"] in region_dict["동북권"]:
        df.loc[idx, "권역"] = "동북권"
    elif row["자치구"] in region_dict["서북권"]:
        df.loc[idx, "권역"] = "서북권"
    elif row["자치구"] in region_dict["서남권"]:
        df.loc[idx, "권역"] = "서남권"
    elif row["자치구"] in region_dict["동남권"]:
        df.loc[idx, "권역"] = "동남권"
        
df["권역"]
```

```python
# 모범 답안

df_pivot = df.pivot_table(index='권역', values=['합계', '세대', '여자', '한국인 계', '등록외국인 계', '65세이상고령자'], aggfunc='sum')
```

```python
import numpy as np

df_pivot = pd.pivot_table(df, index="권역", values=['합계', '세대', '여자', '한국인 계', '등록외국인 계', '65세이상고령자'], aggfunc=np.sum)
df_pivot
```

```python
df_pivot.sort_values(by="합계", ascending=False, inplace=True)
df_pivot
```

```python
df_pivot["고령자비율"] = df_pivot["65세이상고령자"] / df_pivot["합계"] * 100
df_pivot["외국인비율"] = df_pivot["등록외국인 계"] / df_pivot["합계"] * 100
df_pivot["여성비율"] = df_pivot["여자"] / df_pivot["합계"] * 100
df_pivot["세대당인구"] = (df_pivot["합계"] - df_pivot["등록외국인 계"]) / df_pivot["세대"]
df_pivot
```

```python
df_pivot.sort_values(by="외국인비율", inplace=True)
df_pivot
```

- 상관계수(correlation coefficient): 두 변수가 함께 변하는 정도를 -1 ~ +1 범위의 수로 나타낸 것

- 피어슨 상관계수: 칼 피어슨(Karl Pearson)이 개발한 상관계수로, 일반적으로 상관계수라고 하면 피어슨 상관계수를 말함
	- Standard Correlation Coefficient
	- r(상관계수) = X와 Y가 함께 변하는 정도 / X와 Y가 각각 변하는 정도
<br></br>
- 상관계수 행렬(Correlation Matrix): 변수간 상관계수를 보여주는 행렬

```python
# 모범 답안

pivot_corr = df_pivot[['고령자비율', '외국인비율', '여성비율', '세대당인구']].corr()
```

```python
pivot_corr = df_pivot[['고령자비율', '외국인비율', '여성비율', '세대당인구']]
pivot_corr
```

```python
pivot_corr = pivot_corr.corr()
pivot_corr
```
