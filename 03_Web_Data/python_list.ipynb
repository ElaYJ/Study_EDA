# Python List 데이터형 

### 🔰 list 선언

- list형은 대괄호로 생성 

```python
colors = ["red", "blue", "green"]

print(len(colors))
colors[0], colors[1], colors[2]
```

### 🔰 Shallow Copy

```python
b = colors
b
```

```python
b[1] = "black"
b
```

```python
colors
```

### 🔰 Deep Copy

```python
c = colors.copy()
c
```

```python
c[1] = "yellow"
c
```

```python
colors
```

### 🔰 with 반복문

- list형을 반복문(for)에 적용 

```python
for color in colors: 
    print(color)
```

### 🔰 with 조건문

- `in` 키워드로 조건문(if)에 적용 

```python
if "white" in colors:
    print("True")
```

```python
movies = ["라라랜드", "먼 훗날 우리", "어벤저스", "다크나이트"]
print(movies)
```

### 🔰 요소 편집


- `append` : list 제일 뒤에 요소 하나 추가

```python
movies.append("타이타닉")
movies
```

- `pop` : 리스트 제일 뒤부터 자료를 하나씩 삭제 

```python
movies.pop(), movies
```

- `extend` : 제일 뒤에 다른 리스트 자료 추가 

```python
movies.extend(["위대한 쇼맨", "인셉션", "터미네이터", "어벤저스"])
movies
```

- `remove` : 해당 아이템의 첫 번째 자료를 삭제 

```python
movies.remove("어벤저스")
movies
```

- `slicing` : [n:m] n번재 부터 m-1까지 

```python
movies[3:5]
```

```python
favorite_movies = movies[3:5]
favorite_movies
```

- `insert` : 원하는 위치에 자료를 삽입

	- list안에 list를 자료로 추가할 수 있다.

```python
favorite_movies.insert(1, 9.60)
favorite_movies
```

```python
favorite_movies.insert(3, 9.50)
favorite_movies
```

```python
favorite_movies.insert(5, ["마허살라 알리", "린다 카델리아", "비고 모텐슨"])
favorite_movies
```

### 🔰 자료형 확인

- `isinstance` : 자료형이 list인지 확인해 True/False 반환

```python
isinstance(favorite_movies, list)
```

```python
favorite_movies
```

```python
for each_item in favorite_movies:
    if isinstance(each_item, list):
        for nested_item in each_item:
            print("nested_item", nested_item)
    else:
        print("each_item", each_item)
```
