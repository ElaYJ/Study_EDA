- Îã§ÏñëÌïú Ïä§ÌÉÄÏùºÏùÑ ÏßÄÏ†ïÌï† Ïàò ÏûàÎã§.

```python
t = list(range(0, 7)) # t = [0, 1, 2, 3, 4, 5, 6]
y = [1, 4, 5, 8, 9, 5, 3]

def drawGraph():

    plt.figure(figsize=(7, 4))
    plt.plot(
        t,
        y,
        color="red", 
        linestyle="--", # "dashed"  /  "-"Îäî Ïã§ÏÑ†
        marker="o", 
        markerfacecolor="blue",
        markersize=7, 
    )

	# Ï∂ïÏùò Î≤îÏúÑÎ•º ÏßÄÏ†ï (limitÍ∞í)
    plt.xlim([-0.5, 6.5]) 
    plt.ylim([0.5, 9.5])
    plt.show() 
    
drawGraph()
```

#### üî∞ ÏòàÏ†ú3. scatter plot

- scatterÎäî Ï†êÏùÑ ÎøåÎ¶¨ÎìØÏù¥ Í∑∏Î¶¨Îäî Í∑∏Î¶ºÏù¥Îã§.

```python
t = np.array(range(0, 10))
y = np.array([9, 8, 7, 9, 8, 3, 2, 4, 3, 4])
```

```python

def drawGraph():
    plt.figure(figsize=(7, 4))
    plt.scatter(t, y)
    plt.grid(True)
    plt.show()
    
drawGraph()
```

- colorbar ÏÇ¨Ïö©

	- Ï∞∏Í≥†: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.colorbar.html

```python
colormap = t

def drawGraph():
    plt.figure(figsize=(10, 6))
    plt.scatter(t, y, s=150, marker='<', c=colormap)
    plt.colorbar()
    
drawGraph()
```

```python
def drawGraph():
    plt.figure(figsize=(7, 6))
    plt.scatter(t, y, s=50, marker='<', c=colormap)
    plt.colorbar(location='bottom')    
    
drawGraph()
```

```python
def drawGraph():
    plt.figure(figsize=(7, 4))
    plt.scatter(t, y, s=50, marker='<', c=colormap)
    plt.colorbar(orientation='horizontal',fraction=0.05) # shrink=0.5
    
drawGraph()
```
