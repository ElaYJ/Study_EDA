#### 🔰 예제1. 그래프 기초

**삼각함수 그리기**

- np.arange(a, b, s): a부터 b까지 s의 간격 

- np.sin(value)

```python
import numpy as np

t = np.arange(0, 12, 0.01) # 1200개의 값
y = np.sin(t)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, y)
plt.plot(t, np.cos(t))
plt.show()
```

#### 🔰 예제2. 그래프 커스텀

```python
t = np.arange(0, 5, 0.5)
t
```

```python
plt.figure(figsize=(7, 4))
plt.plot(t, t, "r--") # red dash
plt.plot(t, t**2, "bs") # blue square
plt.plot(t, t**3, "g>") # green ▶  /  g^ : green ▲
plt.show()
```

- 다양한 스타일을 지정할 수 있다.

```python
t = list(range(0, 7)) # t = [0, 1, 2, 3, 4, 5, 6]
y = [1, 4, 5, 8, 9, 5, 3]

def drawGraph():

    plt.figure(figsize=(7, 4))
    plt.plot(
        t,
        y,
        color="red", 
        linestyle="--", # "dashed"  /  "-"는 실선
        marker="o", 
        markerfacecolor="blue",
        markersize=7, 
    )

	# 축의 범위를 지정 (limit값)
    plt.xlim([-0.5, 6.5]) 
    plt.ylim([0.5, 9.5])
    plt.show() 
    
drawGraph()
```

#### 🔰 예제3. scatter plot

- scatter는 점을 뿌리듯이 그리는 그림이다.

```python
t = np.array(range(0, 10))
y = np.array([9, 8, 7, 9, 8, 3, 2, 4, 3, 4])
```

```python

def drawGraph():
    plt.figure(figsize=(7, 4))
    plt.scatter(t, y)
    plt.grid(True)
    plt.show()
    
drawGraph()
```

- colorbar 사용

	- 참고: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.colorbar.html

```python
colormap = t

def drawGraph():
    plt.figure(figsize=(10, 6))
    plt.scatter(t, y, s=150, marker='<', c=colormap)
    plt.colorbar()
    
drawGraph()
```

```python
def drawGraph():
    plt.figure(figsize=(7, 6))
    plt.scatter(t, y, s=50, marker='<', c=colormap)
    plt.colorbar(location='bottom')    
    
drawGraph()
```

```python
def drawGraph():
    plt.figure(figsize=(7, 4))
    plt.scatter(t, y, s=50, marker='<', c=colormap)
    plt.colorbar(orientation='horizontal',fraction=0.05) # shrink=0.5
    
drawGraph()
```
