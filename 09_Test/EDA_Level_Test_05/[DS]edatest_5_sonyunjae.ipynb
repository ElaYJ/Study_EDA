```python
def remove_outliers(row):
    for val in row:
        if (val % 10 != 0) | (val >= 100000):
            df_target.drop(row.name, inplace=True)
            return
```

        - 관람가능시간이 6시간 40분과 같이 무환소수(6.6666666666666......6666666667)로 표기될 경우 소숫점 셋째 자리에서 반올림하여 소숫점 둘째 자리까지 표기합니다.

```python
def viewing_hours(row):
    open_h, open_m = map(float, row.평일관람시작시각.split(":"))
    close_h, close_m = map(float, row.평일관람종료시각.split(":"))
    minutes = (close_m - open_m)/60
    hours = (close_h - open_h) + round(minutes, 2)
    df_copy.loc[row.name, '평일관람가능시간'] = 24. if hours > 23 else hours
    
    open_h, open_m = map(float, row.공휴일관람시작시각.split(":"))
    close_h, close_m = map(float, row.공휴일관람종료시각.split(":"))
    minutes = (close_m - open_m)/60
    hours = (close_h - open_h) + round(minutes, 2)
    df_copy.loc[row.name, '공휴일관람가능시간'] = 24. if hours > 23 else hours
    

df_copy.apply(viewing_hours, axis=1)
```
