👉 __수치형 컬럼 형변환__

`df['col'].astype(dtype)`

- 모든 자료형으로 바꿀 수 있다.
- astype은 errors인자를 지원한다. 바꿀수 없을 때 에러를 발생시킬지(‘raise’), 무시할지(‘ignore’)를 결정할수 있다.
	- ‘ignore’는 숫자로 바꿀 수 없는 데이터를 만나면,
	에러만 발생시키지 않을뿐 전체를 아예 바꾸지 않고 원래 데이터를 그대로 돌려준다.

`pd.to_numeric(df['col'])`

- pd.to_numeric(적용할시리즈, errors=“에러처리방식”)
	- errors 인자 인수로 ‘raise’, ‘ignore’뿐만 아니라 astype과는 달리 ‘coerce’를 추가로 지원한다.
	-  ‘coerce’는 바꿀 수 없는 데이터를 만나면 그 값을 강제로 NaN으로 바꾸고 계속 진행한다.

- 숫자형을 바꿀 때 특화된 함수
- 데이터프레임이 아닌 시리즈에만 적용이 되는 함수
- 데이터프레임 전체를 바꾸려면 각각의 열을 시리즈로 뽑아서 반복문으로 바꾸거나 apply를 사용해야 한다.

```python
def viewing_hours(row):
    open_h, open_m = map(float, row.평일관람시작시각.split(":"))
    close_h, close_m = map(float, row.평일관람종료시각.split(":"))
    minutes = (close_m - open_m)/60
    hours = (close_h - open_h) + round(minutes, 2)
    df_copy.loc[row.name, '평일관람가능시간'] = 24. if hours > 23 else hours
    
    open_h, open_m = map(float, row.공휴일관람시작시각.split(":"))
    close_h, close_m = map(float, row.공휴일관람종료시각.split(":"))
    minutes = (close_m - open_m)/60
    hours = (close_h - open_h) + round(minutes, 2)
    df_copy.loc[row.name, '공휴일관람가능시간'] = 24. if hours > 23 else hours
    

df_copy.apply(viewing_hours, axis=1)
df_copy
```
