💡  `apply()` 메서드로 인수함수를 적용해서 바꿀수는 없을까?

---> 값을 찾아 바꾸는게 아니고 인덱스로 찾아야 해서 적용 방법을 찾지 못했음.

└▶ apply() 행 or 열 적용 예제

- axis=1 : 각 행에 대해 함수를 적용

	- row.name --> 행의 인덱스
<br></br>

- axis=0 : 각 열에 대해 함수를 적용

	- col_val.index --> 컬럼 값의 해당 인덱스

```python
import pandas as pd

# 샘플 데이터프레임 생성
data = {
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8],
    'C': [9, 10, 11, 12]
}
df = pd.DataFrame(data)
display(df)

# 각 행 또는 열에 대해 인덱스 값을 함수에 전달하여 적용하는 함수 정의
# 여기서는 간단히 행의 인덱스 값을 10을 곱한 값으로 반환하는 예시를 들었습니다.
def custom_function(row):
    index_value = row.name  # 행의 인덱스 값
    print(index_value)
    return row * index_value

# 각 행에 대해 함수를 적용할 때
result_rows = df.apply(custom_function, axis=1)
print("행에 함수 적용 결과:")
display(result_rows)

# 각 열에 대해 함수를 적용할 때 --> custom_function 사용시 Error 발생 : 컬럼에서는 row.name이 적용되지 않음.
result_columns = df.apply(func=lambda x: x*x.index, axis=0) # (custom_function, axis=1)
print("\n열에 함수 적용 결과:")
display(result_columns)
```

─▶ `pycountry.countries`

<국가 검색 방법>

- countries.get()
	- 일반 검색: pycountry.countries.get(name=country_name) -> 하나의 결과값을 return

- countries.search_fuzzy()
    - fuzzy 검색: pycountry.countries.search_fuzzy(country_name) -> 하나 이상의 결과값을 list형태로 return

```python
from pycountry import countries

for idx, row in df_target.iterrows():
    country_name = row['country']
    try:
        country_code = countries.get(name=country_name)
        df_target.loc[idx, 'code'] = country_code.alpha_2
    except:
        country_code = countries.search_fuzzy(country_name)
        df_target.loc[idx, 'code'] = country_code[0].alpha_2

df_target
```

🚩  __`pd.pivot_table()` 사용__

```python
import pandas as pd
import numpy as np

# 샘플 데이터프레임 생성
data = {
    'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Value': [10, 20, 30, 40, 50, 60],
    'Weight': [1, 2, 3, 4, 5, 6]
}
df_weight = pd.DataFrame(data)

# pivot_table을 사용하여 Category 별로 가중평균을 계산
pivot = pd.pivot_table(
    df_weight, values='Value', index='Category', 
    aggfunc=lambda x: np.average(x, weights=df_weight.loc[x.index, 'Weight'])
)
print(pivot)
"""
OR
"""
pivot = pd.pivot_table(
    df_weight, values='Value', index='Category',
    aggfunc=lambda x: sum(x * df_weight.loc[x.index, 'Weight']) / sum(df_weight.loc[x.index, 'Weight'])
)
print(pivot)
```
