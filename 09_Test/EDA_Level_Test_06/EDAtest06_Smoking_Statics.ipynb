````python
def get_women_ratio(row):
    try:
        return round(((row.전체-row.남자)/(row.여자-row.남자) * 100), 6)
    except:
        return None

df_target['여자비율'] = df_target.apply(get_women_ratio, axis=1)
df_target.head(20)
````

````python
def check(df):
    question_no, points = 3, 10

    condition_dict = {
        'condition01': '여자비율' in df.columns,
        'condition02': (df.여자비율.isna().index != df.여자.isna().index).sum() == 0,
        'condition03': (df.여자비율.apply(lambda x: round(x, 6)).fillna('-') != df.여자비율.fillna('-')).sum() == 0,
        'condition04': df.loc[4, '여자비율'] == 56.783920,
    }

    if sum(condition_dict.values()) == len(condition_dict):
        return True
    else:
        return False

check(df_target)
````

    - 3-2의 df_result '구분1' Column(열) Data마다 어떤 '구분2' Column(열)의 Data가 '여자' 흡연율 증가율이 높은지/낮은지 알아보고자 합니다.
<br></br>
<br></br>
    - 조건1: 결과 DataFrame의 Column(열)명은 ['구분1', '최대', '최소'] 입니다.

    - 조건2: 회귀계수가 양수(=흡연율이 증가중)인 '구분1' Column(열) Data 중 어떤 '구분2' Column(열) Data가 '여자' 흡연율 감소율이 가장 높은지/낮은지 찾아서 결과 DataFrame에 입력하세요.
        - '구분1' 중 '시'는 '구분2' Data 가 1개(서울시)이므로 제외합니다.
        - 만약, '구분1' Column(열) Data 중 증가중인 '구분2' Column(열) Data가 1개라면, 해당 '구분2' Data를 '최대'에 넣고, '최소' Column(열)에는 None값을 입력하세요.
        - 만약, '구분1' Column(열) Data 중 증가중인 '구분2' Column(열) Data가 없다면, 해당 '구분1' Data는 건너뛰세요.(결과 DataFrame에 입력하지 마세요)
<br></br>
    - 조건3: ['구분1']을 기준으로 오름차순으로 순서를 변경(sort)하고 Index를 Reset하세요.
<br></br>
    - 예시(임의의 값을 넣었습니다.)
<br></br>
        <img src="images/06-3-4-001.png" style="margin:auto" width=300/>
<br></br>
    - 조건4: 'df_maxmin' 변수에 결과 DataFrame을 할당하세요.
<br></br>
<br></br>
- 완료 후 결과 dataframe 변수(df_maxmin)를 check_03_04 함수에 입력하여 채점하세요.

````python
def get_title(x):
    return df_result[df_result['여자'] == x]['구분2']
    
def 최대(vals):
    try:
        maxi = max([n for n in vals if n > 0])
        return get_title(maxi)
    except:
        return

def 최소(vals):
    try:
        list = [n for n in vals if n > 0]
        if len(list) > 1:
            return get_title(min(list))
        else:
            return None
    except:
        return

df_maxmin = df_result[['구분1', '여자']].groupby('구분1').aggregate(
    [최대,최소]).droplevel(level=0, axis=1
).reset_index().drop([0,3])
df_maxmin.reset_index(drop=True, inplace=True)
df_maxmin
````
