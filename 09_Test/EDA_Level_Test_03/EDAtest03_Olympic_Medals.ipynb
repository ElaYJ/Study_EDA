```python
import pandas as pd

# 예제 DataFrame 생성
data = {'Gold': [5, 3, 8],
        'Silver': [10, 8, 4],
        'Bronze': [8, 4, 6]}
df = pd.DataFrame(data, index=['USA', 'Canada', 'Germany'])

# 인덱스를 정렬 (sort_index 메서드 사용)
order = ['Gold', 'Silver', 'Bronze']
df_sorted = df.loc[:, order].sort_index(axis=0, key=lambda x: x.map({medal: i for i, medal in enumerate(order)}))

print(df_sorted)
"""
위 코드에서는 sort_index 메서드를 사용하여 DataFrame을 정렬하고,
key 매개변수를 사용하여 'Gold', 'Silver', 'Bronze'의 순서에 따라 정렬했습니다.
axis=0는 행 인덱스를 기준으로 정렬함을 의미합니다.

이 코드에서 enumerate(order)는 'Gold', 'Silver', 'Bronze' 순으로 0, 1, 2의 인덱스를 부여하고,
map({medal: i for i, medal in enumerate(order)})를 사용하여 각 국가의 메달 순서를 매핑하였습니다.
"""
```
