#### `fillna()`

- nanÏùÑ Ï±ÑÏö∞Îäî Ìï®ÏàòÏù¥Îã§.

- Îç∞Ïù¥ÌÑ∞Ïóê NaN Í∞íÏù¥ Ï°¥Ïû¨ÌïúÎã§.</br>
	Ïù¥ NaNÍ∞íÏùÄ ÏóëÏÖÄ ÏÖÄÏù¥ Î≥ëÌï©ÎêòÏñ¥ ÏûàÏñ¥ ÎÇòÌÉÄÎÇú Í∞íÏúºÎ°ú Î≥ëÌï©Îêú Í∞íÏúºÎ°ú ÎèôÏùºÌïòÍ≤å Ï±ÑÏõåÏ§òÏïº ÌïúÎã§.

- Î≥ëÌï© valueÎäî NaN ÏúÑÏóê ÏûàÎäî Í∞íÏù¥Îã§.

- NaNÍ∞íÏùÑ Î≥ëÌï©Îêú valueÎ°ú Ï±ÑÏö∞Í∏∞ ÏúÑÌï¥ fillna() Ìï®ÏàòÎ•º ÏÇ¨Ïö©ÌïúÎã§.

---
- Docstring:

	Fill NA/NaN values using the specified method.

	Parameters
	----------
	- `value` : scalar, dict, Series, or DataFrame

		Value to use to fill holes (e.g. 0), alternately a dict/Series/DataFrame of values </br>
		specifying which value to use for each index (for a Series) or column (for a DataFrame).</br>
		Values not in the dict/Series/DataFrame will not be filled. This value cannot be a list.</br>
	  
	- `method` : {'backfill', 'bfill', 'ffill', None}, default None

		Method to use for filling holes in reindexed Series:</br>

		* ffill: propagate last valid observation forward to next valid.</br>
			--> Ïù¥Ï†Ñ Ïπ∏Ïùò ÎßàÏßÄÎßâ Í∞íÏúºÎ°ú Ï±ÑÏö¥Îã§.

		* backfill / bfill: use next valid observation to fill gap.</br>
			--> Îã§Ïùå Ïπ∏Ïùò Í∞íÏúºÎ°ú Ï±ÑÏö¥Îã§.
	
	- `axis` : {0 or 'index', 1 or 'columns'}

		Axis along which to fill missing values. For 'Series' this parameter is unused and defaults to 0.
	    
	- `inplace` : bool, default False

		If True, fill in-place. Note: this will modify any this will modify any other views on this object
		 (e.r a column in a DataFrame).

- ÏÜåÎ©∏ÏúÑÌóòÏßÄÏàòÎäî?</br>
	65ÏÑ∏ Í≥†Î†πÏù∏Íµ¨ ÎåÄÎπÑ 20-39ÏÑ∏ Ïó¨ÏÑ±Ïù∏Íµ¨ ÎπÑÏ§ëÏúºÎ°ú ÎπÑÏú®Ïù¥ 0.5Ïù¥ÌïòÏù¥Î©¥ 30ÎÖÑ ÎÇ¥ ÏÜåÎ©∏ Í∞ÄÎä•ÏÑ±Ïù¥ ÌÅ¨Îã§Í≥† ÌåêÎã®ÌïúÎã§.

- ÏÜåÎ©∏ÏúÑÌóòÏßÄÏàò = (20-39ÏÑ∏ Ïó¨ÏÑ± Ïù∏Íµ¨Ïàò) / (65ÏÑ∏ Ïù¥ÏÉÅ Ïù∏Íµ¨Ïàò)

- ÏÜåÎ©∏ÏúÑÌóòÏßÄÏàòÍ∞Ä 0.5 ÎØ∏ÎßåÏù¥Î©¥ Ïù∏Íµ¨ ÏÜåÎ©∏ ÏúÑÌóò ÏßÄÏó≠ÏúºÎ°ú Íµ¨Î∂ÑÌïúÎã§.

#### 1> ÎÇòÏù¥ Ïª¨Îüº Ìï©ÏπòÍ∏∞

- 20~39ÏÑ∏ Ïó¨ÏÑ± Ïù∏Íµ¨ÏôÄ 65ÏÑ∏ Ïù¥ÏÉÅ Ïù∏Íµ¨ÏàòÎ•º Íµ¨ÌïòÍ∏∞ ÏúÑÌï¥ ÎÇòÎâòÏñ¥ÏßÑ ÎÇòÏù¥ Ïª¨ÎüºÏùÑ Ìï©Ï≥êÏ§ÄÎã§.

```python
population["20-39ÏÑ∏"] = (
    population["20 - 24ÏÑ∏"]
    + population["25 - 29ÏÑ∏"]
    + population["30 - 34ÏÑ∏"]
    + population["35 - 39ÏÑ∏"]
)

population["65ÏÑ∏Ïù¥ÏÉÅ"] = (
    population["65 - 69ÏÑ∏"]
    + population["70 - 74ÏÑ∏"]
    + population["75 - 79ÏÑ∏"]
    + population["80 - 84ÏÑ∏"]
    + population["85 - 89ÏÑ∏"]
    + population["90 - 94ÏÑ∏"]
    + population["95 - 99ÏÑ∏"]
    + population["100+"]
)

population.tail()
```
