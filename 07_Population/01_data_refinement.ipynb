#### `fillna()`

- nan을 채우는 함수이다.

- 데이터에 NaN 값이 존재한다.</br>
	이 NaN값은 엑셀 셀이 병합되어 있어 나타난 값으로 병합된 값으로 동일하게 채워줘야 한다.

- 병합 value는 NaN 위에 있는 값이다.

- NaN값을 병합된 value로 채우기 위해 fillna() 함수를 사용한다.

---
- Docstring:

	Fill NA/NaN values using the specified method.

	Parameters
	----------
	- `value` : scalar, dict, Series, or DataFrame

		Value to use to fill holes (e.g. 0), alternately a dict/Series/DataFrame of values </br>
		specifying which value to use for each index (for a Series) or column (for a DataFrame).</br>
		Values not in the dict/Series/DataFrame will not be filled. This value cannot be a list.</br>
	  
	- `method` : {'backfill', 'bfill', 'ffill', None}, default None

		Method to use for filling holes in reindexed Series:</br>

		* ffill: propagate last valid observation forward to next valid.</br>
			--> 이전 칸의 마지막 값으로 채운다.

		* backfill / bfill: use next valid observation to fill gap.</br>
			--> 다음 칸의 값으로 채운다.
	
	- `axis` : {0 or 'index', 1 or 'columns'}

		Axis along which to fill missing values. For 'Series' this parameter is unused and defaults to 0.
	    
	- `inplace` : bool, default False

		If True, fill in-place. Note: this will modify any this will modify any other views on this object
		 (e.r a column in a DataFrame).

## 🔰 인구 소멸 위험 지역 조사

- 소멸위험지수는?</br>
	65세 고령인구 대비 20-39세 여성인구 비중으로 비율이 0.5이하이면 30년 내 소멸 가능성이 크다고 판단한다.

- 소멸위험지수 = (20-39세 여성 인구수) / (65세 이상 인구수)

- 소멸위험지수가 0.5 미만이면 인구 소멸 위험 지역으로 구분한다.

#### 1> 나이 컬럼 합치기

- 20~39세 여성 인구와 65세 이상 인구수를 구하기 위해 나뉘어진 나이 컬럼을 합쳐준다.

```python
population["20-39세"] = (
    population["20 - 24세"]
    + population["25 - 29세"]
    + population["30 - 34세"]
    + population["35 - 39세"]
)

population["65세이상"] = (
    population["65 - 69세"]
    + population["70 - 74세"]
    + population["75 - 79세"]
    + population["80 - 84세"]
    + population["85 - 89세"]
    + population["90 - 94세"]
    + population["95 - 99세"]
    + population["100+"]
)

population.tail()
```

#### 2> pivot_table

- 인구 소멸 위험 지역을 조사하기 위해 필요한 값들을 모아 새로운 데이터프레임을 만든다.

```python
popul_pivot = pd.pivot_table(
	data=population,
	index=["광역시도", "시도"],
	columns=["구분"],
	values=["인구수", "20-39세", "65세이상"]
)
popul_pivot
```

#### 3> 소멸 지수 계산

```python
popul_pivot["소멸지수"] = popul_pivot["20-39세", "여자"] / popul_pivot["65세이상", "합계"]
popul_pivot
```

#### 4> 소멸 위기 지역 파악

- "소멸위험지역" 컬럼을 만들어 "소멸지수"가 0.5보다 작은 지역에 True 값을 준다. 

```python
popul_pivot["소멸위기지역"] = popul_pivot["소멸지수"] < 0.5
popul_pivot
```

#### 3> 광역시도 & 시도 정리

- 광역시도의 일반 구와 세종시, 그리고 일반 시의 이름을 정리한다.

```python
sido_name = [None] * len(popul_pivot)
```

```python
popul_pivot[["광역시도", "시도"]].values
```

- 일반 시

	일반 도에 포함된 일반 시나 군의 지역 ID는 "시"를 뺀 지역명으로 한다.

	- row["시도"][:-1] --> 마지막 한 글자를 제외한 전체

	- ex. '강릉시' --> '강릉', '옥천군' --> '옥천'
	
	- 고성군은 강원도와 경상남도에 동일한 이름으로 존재하므로 구분해준다.
	
		- ex. 강원도 '고성군' --> '고성(강원)', 경상남도 '고성군' --> '고성(경남)'

```python
for idx, row in popul_pivot[["광역시도", "시도"]].iterrows():
    print(row["광역시도"][-3:]) # 광역시도의 마지막 세글자
    
    if row["광역시도"][-3:] not in ["광역시", "특별시", "자치시"]:
        if row["시도"] == "고성군":
            if row["광역시도"] == "강원도":
                sido_name[idx] = "고성(강원)"
            else: sido_name[idx] = "고성(경남)"
        else:
            sido_name[idx] = row["시도"][:-1]
        
        print(sido_name[idx]) # 마지막 한 글자 제외
```

#### 4> 통합 함수

```python
def find_key_by_value(dictionary, value):
    for key, val in dictionary.items():
        if value in val:
            return key
    return None  # 값이 없는 경우

for idx, row in popul_pivot[["광역시도", "시도"]].iterrows():
    
    if row["광역시도"][-3:] not in ["광역시", "특별시", "자치시"]:
        # 행정구
        if row["시도"][-1:] == "구":
            key = find_key_by_value(admin_gu, row["시도"])            
            if len(key) > 0:
                if len(row["시도"]) == 2:
                    sidoID[idx] = key + " " + row["시도"]                    
                elif row["시도"] in ["마산합포구", "마산회원구"]:
                    sidoID[idx] = key + " " + row["시도"][2:-1]                
                else:
                    sidoID[idx] = key + " " + row["시도"][:-1]
        # 고성군
        elif row["시도"] == "고성군":
            if row["광역시도"] == "강원도":
                sidoID[idx] = "고성(강원)"
            else: sidoID[idx] = "고성(경남)"
        # 일반시
        else:
            sidoID[idx] = row["시도"][:-1]
            
    # 세종특별자치시
    elif row["광역시도"] == "세종특별자치시":
        sidoID[idx] = "세종"
        
    # 특별시/광역시 일반구
    else:
        if len(row["시도"]) == 2: # 서구, 중구, 북구, ...
            sidoID[idx] = row["광역시도"][:2] + " " + row["시도"]
        else:
            sidoID[idx] = row["광역시도"][:2] + " " + row["시도"][:-1]
```
